```python
import os
import re
import zipfile
import glob
import difflib
import cv2
import torch
import time
import numpy as np
from paddleocr import PaddleOCR
from transformers import Qwen2VLForConditionalGeneration, AutoProcessor
from PIL import Image
from io import BytesIO
from fastapi import FastAPI, UploadFile, File, Form, HTTPException
from fastapi.responses import JSONResponse
import uvicorn
import nest_asyncio
import threading
from pyngrok import ngrok
from ipywidgets import HTML
from IPython.display import display
import pandas as pd
from sklearn.preprocessing import binarize
import matplotlib.pyplot as plt
from typing import List
import requests
from langchain.chains import GraphQAChain
from langchain_community.graphs.networkx_graph import NetworkxEntityGraph
from langchain_together import ChatTogether
from langchain.prompts import PromptTemplate
import asyncio
import json

CROPPED_DIR = "/content/drive/MyDrive/cropped_images"
EXCEL_FILE = "/content/drive/MyDrive/diagnosis_data.xlsx"
ABBREVIATION_FILE_PATH = '/content/dataset.txt'
gml_file_url = "https://raw.githubusercontent.com/kartikey-codes/datasets/refs/heads/main/medical_terms_abbreviations.txt"
if not os.path.exists(ABBREVIATION_FILE_PATH):
    print("Downloading abbrevition file...")
    response = requests.get(gml_file_url)
    with open(ABBREVIATION_FILE_PATH, 'wb') as f:
        f.write(response.content)


from google.colab import drive
drive.mount('/content/drive')


nest_asyncio.apply()


asyncio.set_event_loop_policy(asyncio.DefaultEventLoopPolicy())

app = FastAPI()

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize OCR
ocr_device = 'gpu' if torch.cuda.is_available() else 'cpu'
ocr = PaddleOCR(use_angle_cls=True, lang='en', device=ocr_device)


model = Qwen2VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2-VL-2B-Instruct", torch_dtype="auto", device_map="auto"
).to(device)
processor = AutoProcessor.from_pretrained("Qwen/Qwen2-VL-2B-Instruct")


def load_abbreviations(file_path):
    abbreviations_dict = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        for line in lines:
            parts = line.strip().split(' ', 1)
            if len(parts) == 2:
                abbreviation, full_form = parts
                abbreviations_dict[abbreviation.upper()] = full_form
    return abbreviations_dict

def clean_text(text):
    # Remove symbols and unnecessary spaces
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Keep only alphanumeric characters and spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

    # Remove units of time (e.g., "36 weeks", "2 days", "5 hours", "week 36")
    cleaned_text = re.sub(r'\b\d+\s*(weeks?|days?|hours?|minutes?|seconds?|min?|month?|months?|hr?)\b', '', cleaned_text)
    cleaned_text = re.sub(r'\b(weeks?|days?|hours?|minutes?|seconds?|min?|month?|months?|hr?)\s*\d+\b', '', cleaned_text)

    return cleaned_text

def replace_abbreviations(sentence, abbreviations_dict):
    cleaned_sentence = clean_text(sentence)
    words = cleaned_sentence.split()
    updated_words = []
    for word in words:
        stripped_word = re.sub(r'[^\w\s]', '', word)
        full_form = abbreviations_dict.get(stripped_word.upper(), None)
        if full_form:
            updated_word = word.replace(stripped_word, full_form)
            updated_words.append(updated_word)
        else:
            updated_words.append(word)
    return ' '.join(updated_words)


def read_image_from_bytes(image_bytes):
    image = Image.open(BytesIO(image_bytes)).convert("RGB")
    return image

def crop_provisional_diagnosis(img, ocr_results):
    target_text = "provisional diagnosis"

    for line in ocr_results:
        for word_info in line:
            text = word_info[1][0].lower()

            # Fuzzy matching to check similarity between detected text and 'provisional diagnosis'
            similarity = difflib.SequenceMatcher(None, text, target_text).ratio() * 100

            if similarity >= 90:
                bbox = word_info[0]
                x_min = int(min([point[0] for point in bbox])) + 100
                y_min = int(min([point[1] for point in bbox])) - 20
                x_max = int(max([point[0] for point in bbox])) + 1000
                y_max = int(max([point[1] for point in bbox])) + 30

                cropped_img = img[y_min:y_max, x_min:x_max]
                return cropped_img
    return None

def extract_provisional_diagnosis(image_bytes):
    start_time = time.time()
    # Read image from bytes
    img = cv2.imdecode(np.frombuffer(image_bytes, np.uint8), cv2.IMREAD_COLOR)
    if isinstance(img, Image.Image):
        img = np.array(img)
    result = ocr.ocr(img, cls=True)
    cropped_img = crop_provisional_diagnosis(img, result)
    elapsed_time = time.time() - start_time
    print(f"Time taken for extract_provisional_diagnosis: {elapsed_time:.2f} seconds")
    return cropped_img

def process_vision_info(messages):
    image_inputs = []
    for message in messages:
        for content in message["content"]:
            if content["type"] == "image":
                image_inputs.append(content["image"])
    return image_inputs, None


def thinning(image):
    kernel = np.ones((3, 3), np.uint8)
    eroded_img = cv2.erode(image, kernel, iterations=1)
    dilated_img = cv2.dilate(eroded_img, kernel, iterations=1)
    return dilated_img

# Function to clean the image
def clean_image(image, scale_factor=2.5):

    if isinstance(image, Image.Image):
        image = np.array(image)

    height, width = image.shape[:2]
    new_size = (int(width * scale_factor), int(height * scale_factor))
    image = cv2.resize(image, new_size, interpolation=cv2.INTER_CUBIC)

    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    # Apply the thinning process
    cleaned_image = thinning(gray_image)

    # Convert back to PIL Image
    cleaned_image = Image.fromarray(cleaned_image)
    return cleaned_image

def extract_text_from_image(image):
    start_time = time.time()
    cleaned_image = clean_image(image)

    messages = [
        {
            "role": "user",
            "content": [
                {"type": "image", "image": cleaned_image},
                {"type": "text", "text": "Extract provisional diagnosis from the image, and correct any inappropriate word based on contextual understanding"}
            ],
        }
    ]

    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    image_inputs, _ = process_vision_info(messages)
    inputs = processor(text=[text], images=image_inputs, padding=True, return_tensors="pt").to(device)
    generated_ids = model.generate(**inputs, max_new_tokens=128)
    generated_ids_trimmed = [
        out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]
    output_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )
    print(output_text)
    elapsed_time = time.time() - start_time
    print(f"Time taken for extract_text_from_image: {elapsed_time:.2f} seconds")

    return output_text[0] if output_text else ""

def refine_extracted_text(extracted_text):
    start_time = time.time()

    messages = [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": extracted_text},
                {"type": "text", "text": "Refine the extracted diagnosis. Correct spelling mistakes and ensure that any misrecognized medical terms are corrected based on context."}
            ],
        }
    ]

    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], padding=True, return_tensors="pt").to(device)
    generated_ids = model.generate(**inputs, max_new_tokens=128)
    generated_ids_trimmed = [
        out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]
    refined_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )
    elapsed_time = time.time() - start_time
    print(f"Time taken for refine_extracted_text: {elapsed_time:.2f} seconds")
    return refined_text[0] if refined_text else ""

# Download GML file if not already present
gml_file_url = "https://raw.githubusercontent.com/kartikey-codes/datasets/refs/heads/main/icd10_graph.gml"
gml_file_path = "/content/icd10_graph.gml"
if not os.path.exists(gml_file_path):
    print("Downloading GML file...")
    response = requests.get(gml_file_url)
    with open(gml_file_path, 'wb') as f:
        f.write(response.content)

graph = NetworkxEntityGraph.from_gml(gml_file_path)

chat_model = ChatTogether(
    together_api_key="019d5b390bf9b379deefbc22ed4cb09750a72e",  # Replace with your actual API key
    model="meta-llama/Llama-3-70b-chat-hf",
)


qa_generation_template = """You are an AI Agent specialized in retrieving relevant information about ICD-10 coding.
The user has provided a provisional diagnosis, and your task is to suggest the most appropriate ICD-10 code based **only** on the graph context provided.
The codes and their descriptions are provided as [c: code, d: description, ...].

Instructions:
- Analyze the provisional diagnosis provided by the user.
- **Return only the most relevant code and its description from the context.**
- **Your answer must be based solely on the codes and descriptions provided in the context. Do not use any external information.**
- **Return your answer in JSON format as: {{"icd10_code": "XXX", "description": "YYY"}}**

Provisional Diagnosis: {question}
Context: {context}

Answer:
"""

entity_prompt_template = """You are tasked with extracting relevant ICD-10 codes from a graph based on a provisional diagnosis.
Each node represents an ICD-10 code or medical term. Your goal is to list out all codes and their descriptions that are relevant to the provisional diagnosis, restricting your answer **only** to what exists in the graph.
If the provisional diagnosis contains terms that closely match words in the context, match those words to the context.

Here are the extracted triples from the graph:
{context}

Please provide the extracted codes and their descriptions, along with relationships between them.

Extracted Entities:
"""

qa_generation_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=qa_generation_template
)

entity_prompt = PromptTemplate(
    input_variables=["context"],
    template=entity_prompt_template
)


chain = GraphQAChain.from_llm(
    llm=chat_model,
    graph=graph,
    qa_prompt=qa_generation_prompt,
    entity_prompt=entity_prompt,
    verbose=True
)

def get_icd10_code(extracted_text):
    try:

        response = chain.invoke({
            "query": extracted_text
        })

        print("Chain Response:", response)

        result_text = response.get('result', '')


        match = re.search(r'\{.*\}', result_text, re.DOTALL)
        if match:
            json_str = match.group(0)

            icd10_data = json.loads(json_str)
            icd10_code = icd10_data.get('icd10_code', '')
            description = icd10_data.get('description', '')
        else:

            icd10_code = ''
            description = ''

        return {
            "icd10_code": icd10_code,
            "description": description,
            "extracted_text": extracted_text
        }
    except Exception as e:
        print(f"Error in get_icd10_code: {e}")
        return None


@app.post("/extract_diagnosis")
async def extract_diagnosis(file: UploadFile = File(...)):
    try:

        image_bytes = await file.read()

        cropped_img = extract_provisional_diagnosis(image_bytes)
        if cropped_img is not None:

            cropped_image_pil = Image.fromarray(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))

            extracted_diagnosis = extract_text_from_image(cropped_image_pil)
            abbreviations_dict = load_abbreviations(ABBREVIATION_FILE_PATH)
            extracted_abbreviated_diagnosis = replace_abbreviations(extracted_diagnosis, abbreviations_dict)

            refined_diagnosis = refine_extracted_text(extracted_diagnosis)

            icd10_result = get_icd10_code(refined_diagnosis)
            if icd10_result is not None:

                data = {
                    "Extracted Text": extracted_abbreviated_diagnosis,
                    "Refined Diagnosis": refined_diagnosis,
                    "ICD-10 Code": icd10_result.get("icd10_code"),
                    "Description": icd10_result.get("description")
                }

                if os.path.exists(EXCEL_FILE):
                    df = pd.read_excel(EXCEL_FILE)
                else:
                    df = pd.DataFrame(columns=["Extracted Text", "Refined Diagnosis", "ICD-10 Code", "Description"])
                new_entry = pd.DataFrame([data])
                df = pd.concat([df, new_entry], ignore_index=True)
                df.to_excel(EXCEL_FILE, index=False)

                return data
            else:
                return JSONResponse(content={"error": "Could not retrieve ICD-10 code"}, status_code=500)
        else:
            return JSONResponse(content={"error": "Could not extract provisional diagnosis section"}, status_code=400)
    except Exception as e:
        return JSONResponse(content={"error": str(e)}, status_code=500)

@app.post("/process-folder/")
async def process_folder(folder_path: str = Form(...)):
    try:

        os.makedirs(CROPPED_DIR, exist_ok=True)


        image_files = glob.glob(os.path.join(folder_path, "*.jpg")) + \
                      glob.glob(os.path.join(folder_path, "*.jpeg")) + \
                      glob.glob(os.path.join(folder_path, "*.png"))

        if not image_files:
            raise HTTPException(status_code=404, detail="No images found in the folder.")

        # List to hold processing results
        processing_results = []
=
        if os.path.exists(EXCEL_FILE):
            df = pd.read_excel(EXCEL_FILE)
        else:
            df = pd.DataFrame(columns=["Image Name", "Extracted Text", "Refined Diagnosis", "ICD-10 Code", "Description"])

        for image_file in image_files:
            with open(image_file, 'rb') as f:
                image_bytes = f.read()

            cropped_image = extract_provisional_diagnosis(image_bytes)

            if cropped_image is None:
                processing_results.append({
                    "image_name": os.path.basename(image_file),
                    "error": "Could not extract provisional diagnosis section"
                })
                continue

            cropped_image_pil = Image.fromarray(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))

            extracted_text = extract_text_from_image(cropped_image_pil)
            abbreviations_dict = load_abbreviations(ABBREVIATION_FILE_PATH)
            extracted_abbreviated_diagnosis = replace_abbreviations(extracted_text, abbreviations_dict)


            refined_diagnosis = refine_extracted_text(extracted_text)


            icd10_result = get_icd10_code(refined_diagnosis)

            new_entry = {
                "Image Name": os.path.basename(image_file),
                "Extracted Text": extracted_abbreviated_diagnosis,
                "Refined Diagnosis": refined_diagnosis,
                "ICD-10 Code": icd10_result.get("icd10_code") if icd10_result else None,
                "Description": icd10_result.get("description") if icd10_result else None
            }

            df = pd.concat([df, pd.DataFrame([new_entry])], ignore_index=True)

            processing_results.append(new_entry)

        df.to_excel(EXCEL_FILE, index=False)

        return JSONResponse(content={"processed_images": processing_results}, status_code=200)

    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

def run():
    uvicorn.run(app, host='0.0.0.0', port=8000, loop="asyncio")

threading.Thread(target=run, daemon=True).start()

ngrok_tunnel = ngrok.connect(8000)
print(f"Public URL: {ngrok_tunnel.public_url}")

t = HTML(
    value="Server is running...",
    description='',
    style={'description_width': 'initial'},
)
display(t)
```
